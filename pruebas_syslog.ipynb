```python
def genSyslogEntry(lines):

    etr ={} 
    etr[parm[0]] = lines[:15] #stime
    etr[parm[1]] = lines[15:].split()[0]

    ori = lines[15:].split()
    for i in range(1,len(ori)):
        if '=' in ori[i] != -1:
            if ori[i].find('"') != -1:
                ori[i] = ori[i].replace('"','')
                k = 1
#                while True and not (ori[i].split('=')[1]=='NA') and k < 10 and i<len(ori)-1 and ori[i].find('=') == -1:
                #print(ori[i]+' i '+str(i)+' len '+ str(len(ori))+' =? '+str(ori[i].find('=')))
                while (k < 10) and (i<len(ori)-1) and (ori[i+k].find('=') == -1):
                   # print(ori[i+k])
                    
                    ori[i] = ori[i]+' '+ori[i+k]
                    if ori[i].find('"')!=-1:
                        ori[i] = ori[i].replace('"','')
                        break
                    k = k+1
            try:
                j = parm.index(ori[i].split('=')[0])
                etr[ori[i].split('=')[0]] = ori[i].split('=')[1]
            except:
                print(str(i)+' parámetro no encontrado  '+ori[i].split('=')[0])
            
    return etr
```
